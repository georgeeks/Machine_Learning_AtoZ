```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset=pd.read_csv('Mall_Customers.csv')
```

```python
dataset.head()
```

```python
X= dataset.iloc[:,[3,4]].values
```

```python
# dendrogram to find the optimal number of clusters 
import scipy.cluster.hierarchy as sch
dendrogram=sch.dendrogram(sch.linkage(X, method='ward'))  # ward method  to minimise the variance within each cluster
plt.title('Dendrogram')
plt.xlabel('Clusters')
plt.ylabel('Euclidean distance')
plt.show()
```

```python
# fitting HC to the mall dataset 
from sklearn.cluster import AgglomerativeClustering
hc= AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
y_hc= hc.fit_predict(X)
```

```python
# Vis
plt.scatter(x=X[y_hc==0, 0], y=X[y_hc==0,1], s=30,c='red', label='cluster1')  # choose the first columns of X where y_kmeans=0
plt.scatter(x=X[y_hc==1, 0], y=X[y_hc==1,1], s=30,c='blue', label='cluster2') # `s` to specify the size of each point
plt.scatter(x=X[y_hc==2, 0], y=X[y_hc==2,1], s=30,c='green', label='cluster3')
plt.scatter(x=X[y_hc==3, 0], y=X[y_hc==3,1], s=30,c='cyan', label='cluster4')
plt.scatter(x=X[y_hc==4, 0], y=X[y_hc==4,1], s=30,c='magenta', label='cluster5')

plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
